```python
import pandas as pd
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
%matplotlib inline

dane= pd.read_csv("reviews_courses.csv", parse_dates=["Timestamp"])
```

```python
dane
```

```python
dane["Day"]=dane['Timestamp'].dt.date
day_average=dane.groupby(['Day']).mean(numeric_only=True)
day_average.head()
x=day_average.index
y=day_average['Rating']
plt.figure(
    figsize=(25,10)
)
plt.plot(x,y)
```

```python
dane["Week"]=dane["Timestamp"].dt.strftime("%Y-%U")
week_avg=dane.groupby(["Week"]).mean(numeric_only=True)
week_avg.head()
plt.figure(
    figsize=(25,10)
)
plt.plot(week_avg.index,week_avg["Rating"])
```

```python
dane["Month"]=dane["Timestamp"].dt.strftime('%Y-%m')
month_avg=dane.groupby(dane["Month"]).mean(numeric_only=True)
plt.figure(
    figsize=(5,10)
)
plt.plot(month_avg.index,month_avg["Rating"])
```

```python
month_average_course=dane.groupby(['Month','Course Name']).mean(numeric_only=True).unstack()
month_average_course.head()
plt.figure(
    figsize=(25,10)
)
plt.plot(month_average_course.index,month_average_course["Rating"])
```

```python
month_average_course=dane.groupby(['Month','Course Name'])['Rating'].count().unstack()
month_average_course.head()
plt.figure(
    figsize=(25,10)
)
plt.plot(month_average_course.index,month_average_course)
```
